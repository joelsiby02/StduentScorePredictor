```python
# Group the data by socioeconomic status (SES)
Group_data_SES = df.groupby('parental level of education')

# Calculate the mean scores for math, reading, and writing among high-SES and low-SES students
SES_scores = Group_data_SES[['math score', 'reading score', 'writing score']].mean()
high_SES_scores = SES_scores.loc[['master\'s degree', 'bachelor\'s degree'], :]
low_SES_scores = SES_scores.loc[['some college', 'associate\'s degree', 'high school', 'some high school'], :]
# Print the mean scores for math, reading, and writing among high-SES and low-SES students
print('Mean scores for math, reading, and writing among high-SES students:\n', high_SES_scores)
print('Mean scores for math, reading, and writing among low-SES students:\n', low_SES_scores)
```

```python
# Plot for high-SES students
fig, axs = plt.subplots(ncols=2, figsize=(10,5))

high_SES_scores.plot(ax=axs[0], kind='bar', y=['math score', 'reading score', 'writing score'], alpha=0.5)
axs[0].set_title('Mean Scores for High-SES Students')
axs[0].set_ylabel('Mean Score')
axs[0].set_xlabel('Parental Level of Education')

# Plot for low-SES students
low_SES_scores.plot(ax=axs[1], kind='bar', y=['math score', 'reading score', 'writing score'], alpha=0.5)
axs[1].set_title('Mean Scores for Low-SES Students')
axs[1].set_ylabel('Mean Score')
axs[1].set_xlabel('Parental Level of Education')

# Compare high-SES and low-SES students
fig, axs = plt.subplots(ncols=2, figsize=(10,5))

high_SES_scores.plot(ax=axs[0], kind='bar', y=['math score', 'reading score', 'writing score'], alpha=0.5)
low_SES_scores.plot(ax=axs[0], kind='bar', y=['math score', 'reading score', 'writing score'], alpha=0.5)
axs[0].set_title('Mean Scores Comparison by SES')
axs[0].set_ylabel('Mean Score')
axs[0].set_xlabel('Parental Level of Education')
axs[0].legend(['High-SES', 'Low-SES'])

high_SES_scores.T.plot(ax=axs[1], kind='hist', bins=20, alpha=0.5)
low_SES_scores.T.plot(ax=axs[1], kind='hist', bins=20, alpha=0.5)
axs[1].set_title('Histogram of Mean Scores by SES')
axs[1].set_xlabel('Mean Score')
axs[1].legend(['High-SES', 'Low-SES'])

plt.tight_layout()
plt.show()
```

### Insights: we can see that the bars for the low level of education are higher than the bars for the high level of education. This means that on average, students whose parents have a low level of education scored higher in all three subjects than students whose parents have a high level of education.
